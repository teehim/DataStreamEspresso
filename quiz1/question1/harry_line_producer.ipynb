```python
import time
from kafka import KafkaProducer
```

```python
producer = KafkaProducer(bootstrap_servers=['ec2-13-229-46-113.ap-southeast-1.compute.amazonaws.com:9092'])
```

```python
with open('./data/book.txt', 'r', newline='', encoding='utf-8') as file:
    for line in file.readlines():
        producer.send('espresso-line-stream', line.strip().encode('utf-8'))
        time.sleep(1)
```
